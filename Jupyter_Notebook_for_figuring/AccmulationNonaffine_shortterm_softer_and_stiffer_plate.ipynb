```python
condition_ = '80G'

c_stiffer_ = np.load('data/coordinate_data/c_former_of_%s_%s_from_%d_to_%d.npy' % \
                 (condition_, 'slip', events_80G_[n_80G_][0][0], events_80G_[n_80G_][1][0])   )
D2min_stiffer_ = np.load('data/nonaffine_data/nonaffine_of_%s_%s_from_%d_to_%d.npy'% \
                        (condition_, 'slip', events_80G_[n_80G_][0][0], events_80G_[n_80G_][1][0]))
# plot smaller slip

plt.rc('font', family = 'Arial')
plt.rcParams['font.size'] = 8
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
fig = plt.figure(figsize=(7, 1))
plt.scatter(c_stiffer_[:,0] - c_stiffer_[:,0].min(),c_stiffer_[:,1] - c_stiffer_[:,1].min(),c = D2min_stiffer_, s = 1, cmap = 'Greys', vmin = 0, vmax = 0.004)
cb = plt.colorbar()
cb. set_label(label = '$\\mathit{D}$'+ '$^\\mathrm{2}$' + '$_\\mathrm{min}$' + ' ' + '(mm$^\\mathrm{2}$)', fontsize = 8)

plt.xlabel('$\\mathit{x}$ (mm, from %d ms to %d ms)' % (time_80G_[events_80G_[n_80G_][0][0]], time_80G_[events_80G_[n_80G_][1][0]]), fontsize = 8)


plt.ylabel('$\\mathit{y}$ (mm)',fontsize = 8)

plt.savefig('notebook_plot/nonaffine_stiffer_plate.svg', dpi=600, format='svg')
plt.show()
```

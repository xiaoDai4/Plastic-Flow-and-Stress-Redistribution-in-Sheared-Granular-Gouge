```python
# import data and calculate CCDF of nonaffine of slips 
# under different normal stress
# read nonaffine
# P12
D2min_P12_ = np.array([0.1])
for i_ in range(len(events_P12_)):
    condition_ = 'P12'
    if i_ == 0:
        continue
    D2min_ = np.load('data/nonaffine_data/nonaffine_of_%s_%s_from_%d_to_%d.npy'% \
                            (condition_, 'stick', events_P12_[i_-1][1][0], events_P12_[i_][0][0]))
    D2min_P12_ = np.concatenate((D2min_P12_, D2min_ / D2min_.mean()), axis = 0)
D2min_P12_ = np.delete(D2min_P12_, 0)
print('Read P12 done!')

# P20
D2min_P20_ = np.array([0.1])
for i_ in range(len(events_P20_)):
    condition_ = 'P20'
    if i_ == 0:
        continue
    D2min_ = np.load('data/nonaffine_data/nonaffine_of_%s_%s_from_%d_to_%d.npy'% \
                            (condition_, 'stick', events_P20_[i_-1][1][0], events_P20_[i_][0][0]))
    D2min_P20_ = np.concatenate((D2min_P20_, D2min_ / D2min_.mean()), axis = 0)
D2min_P20_ = np.delete(D2min_P20_, 0)
print('Read P20 done!')

# P28
D2min_P28_ = np.array([0.1])
for i_ in range(len(events_P28_)):
    condition_ = 'P28'
    if i_ == 0:
        continue
    D2min_ = np.load('data/nonaffine_data/nonaffine_of_%s_%s_from_%d_to_%d.npy'% \
                            (condition_, 'stick', events_P28_[i_-1][1][0], events_P28_[i_][0][0]))
    D2min_P28_ = np.concatenate((D2min_P28_, D2min_  / D2min_.mean()), axis = 0)
D2min_P28_ = np.delete(D2min_P28_, 0)
print('Read P28 done!')

# P36
D2min_P36_ = np.array([0.1])
for i_ in range(len(events_P36_)):
    condition_ = 'P36'
    if i_ == 0:
        continue
    D2min_ = np.load('data/nonaffine_data/nonaffine_of_%s_%s_from_%d_to_%d.npy'% \
                            (condition_, 'stick', events_P36_[i_-1][1][0], events_P36_[i_][0][0]))
    D2min_P36_ = np.concatenate((D2min_P36_, D2min_  / D2min_.mean()), axis = 0)
D2min_P36_ = np.delete(D2min_P36_, 0)
print('Read P36 done!')

# P44
D2min_P44_ = np.array([0.1])
for i_ in range(len(events_P44_)):
    condition_ = 'P44'
    if i_ == 0:
        continue
    D2min_ = np.load('data/nonaffine_data/nonaffine_of_%s_%s_from_%d_to_%d.npy'% \
                            (condition_, 'stick', events_P44_[i_-1][1][0], events_P44_[i_][0][0]))
    D2min_P44_ = np.concatenate((D2min_P44_, D2min_ / D2min_.mean() ), axis = 0)
D2min_P44_ = np.delete(D2min_P44_, 0)
print('Read P44 done!')

# Get the CCDF


# P12
n_,bins_, patches_ = plt.hist(D2min_P12_ , cumulative = True, bins = 1000, )
centers_ = np.zeros(len(n_))
for i_ in range(len(n_)):
    centers_[i_] = (bins_[i_] + bins_[i_ + 1]) / 2   
centers_P12_ = np.copy(centers_)
CCDF_P12_ = 1 - (n_  / len(D2min_P12_))

# P20
n_,bins_, patches_ = plt.hist(D2min_P20_ , cumulative = True, bins = 1000, )
centers_ = np.zeros(len(n_))
for i_ in range(len(n_)):
    centers_[i_] = (bins_[i_] + bins_[i_ + 1]) / 2   
centers_P20_ = np.copy(centers_)
CCDF_P20_ = 1 - (n_  / len(D2min_P20_))

# P28
n_,bins_, patches_ = plt.hist(D2min_P28_ , cumulative = True, bins = 1000, )
centers_ = np.zeros(len(n_))
for i_ in range(len(n_)):
    centers_[i_] = (bins_[i_] + bins_[i_ + 1]) / 2   
centers_P28_ = np.copy(centers_)
CCDF_P28_ = 1 - (n_  / len(D2min_P28_))

# P36
n_,bins_, patches_ = plt.hist(D2min_P36_ , cumulative = True, bins = 1000,)
centers_ = np.zeros(len(n_))
for i_ in range(len(n_)):
    centers_[i_] = (bins_[i_] + bins_[i_ + 1]) / 2   
centers_P36_ = np.copy(centers_)
CCDF_P36_ = 1 - (n_  / len(D2min_P36_))

# P44
n_,bins_, patches_ = plt.hist(D2min_P44_, cumulative = True, bins = 1000, )
centers_ = np.zeros(len(n_))
for i_ in range(len(n_)):
    centers_[i_] = (bins_[i_] + bins_[i_ + 1]) / 2   
centers_P44_ = np.copy(centers_)
CCDF_P44_ = 1 - (n_  / len(D2min_P44_))
```

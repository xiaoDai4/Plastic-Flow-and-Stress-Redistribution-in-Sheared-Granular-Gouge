```python
# define the number of event
n_large_ = 49 
n_small_ = 41
n_long_ = 41
n_short_ = 49
```

```python
# import data and calculate CCDF of nonaffine in slips
# read nonaffine
condition_ = '4G'
D2min_smaller_ = np.load('data/nonaffine_data/nonaffine_of_%s_%s_from_%d_to_%d.npy'% \
                        (condition_, 'slip', events_4G_[n_small_][0][0], events_4G_[n_small_][1][0]))
D2min_larger_ = np.load('data/nonaffine_data/nonaffine_of_%s_%s_from_%d_to_%d.npy'% \
                        (condition_, 'slip', events_4G_[n_large_][0][0], events_4G_[n_large_][1][0]))

# Get the CCDF


# smaller
n_,bins_, patches_ = plt.hist(D2min_smaller_ / D2min_smaller_.mean(), cumulative = True, bins = 100)
centers_ = np.zeros(len(n_))
for i_ in range(len(n_)):
    centers_[i_] = (bins_[i_] + bins_[i_ + 1]) / 2   
centers_small_ = np.copy(centers_)
CCDF_small_ = 1 - (n_  / len(D2min_smaller_))

# larger
n_,bins_, patches_ = plt.hist(D2min_larger_ / D2min_larger_.mean(), cumulative = True, bins = 100)
centers_ = np.zeros(len(n_))
for i_ in range(len(n_)):
    centers_[i_] = (bins_[i_] + bins_[i_ + 1]) / 2   
centers_large_ = np.copy(centers_)
CCDF_large_ = 1 - (n_  / len(D2min_larger_))



```

```python
# plot slip
plt.figure(figsize=(3.35,2.52))
plt.rcParams['font.size'] = 9
plt.rc('font', family = 'Arial')
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'

plt.axes(yscale = "log", xscale = "log",)

plt.plot(centers_small_,  \
         CCDF_small_,\
         linewidth = 1, color = 'darkorange', label = 'Small slip')



plt.plot(centers_large_,  \
         CCDF_large_,\
         linewidth = 1, color = 'r', label = 'Large slip')

plt.legend( fontsize = 8)
plt.xlim(10**0, 10**2)
plt.ylim([10**-3, 1])
# plt.yticks([0, 0.5, 1, 1.5, 2, 2.5])
# plt.xticks([0, 0.025, 0.05, 0.075, 0.1])
plt.legend( fontsize = 7)
plt.xlabel('$\\mathit{D}$'+ '$^\\mathrm{2}$' + '$_\\mathrm{min}$' + ' $\\mathrm{/ <}$' + '$\\mathit{D}$'+ '$^\\mathrm{2}$' + '$_\\mathrm{min}$' + '$\\mathrm{>}$', fontsize = 10)
plt.ylabel('CCDF', fontsize = 10)
plt.tight_layout()
plt.savefig('notebook_plot/CCDF_of_nonaffine_in_slip_examples_normalised.svg', dpi=600, format='svg')
```

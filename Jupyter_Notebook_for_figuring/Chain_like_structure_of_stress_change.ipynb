```python
# define the number of event
n_large_ = 49 
n_small_ = 41
n_long_ = 41
n_short_ = 49
# define the percentile to extract the chain like structure
percent_ = 95
```

```python
# get the stress of small slip

condition_ = '4G'

c_smaller_ = np.load('data/coordinate_data/c_former_of_%s_%s_from_%d_to_%d.npy' % \
                 (condition_, 'slip', events_4G_[n_small_][0][0], events_4G_[n_small_][1][0])   )
stress_small_former_ = np.load('data/stress_data/stresses_at_%d_%s.npy'% \
                        (events_4G_[n_small_][0][0], condition_) )
stress_small_later_ =  np.load('data/stress_data/stresses_at_%d_%s.npy'% \
                        (events_4G_[n_small_][1][0], condition_) )
tau_small_former_ = np.zeros(stress_small_former_.shape[0])
tau_small_later_ = np.zeros(stress_small_former_.shape[0])
# count_ = 0
for i_ in range(stress_small_former_.shape[0]):

    c_ = np.array([[stress_small_former_[i_][0], stress_small_former_[i_][2]],
              [stress_small_former_[i_][2], stress_small_former_[i_][1]]])
    eig, vector = np.linalg.eig(c_)
    maxarg_ = np.argmax(eig)
    minarg_ = np.argmin(eig)
    tau_small_former_[i_] = (eig[maxarg_] - eig[minarg_])/2
    
    c_ = np.array([[stress_small_later_[i_][0], stress_small_later_[i_][2]],
              [stress_small_later_[i_][2], stress_small_later_[i_][1]]])
    eig, vector = np.linalg.eig(c_)
    maxarg_ = np.argmax(eig)
    minarg_ = np.argmin(eig)
    tau_small_later_[i_] = (eig[maxarg_] - eig[minarg_])/2
stress_change_small_ =  tau_small_later_ - tau_small_former_
```

```python
# get the chain of small slip

# get the marker
markers_small_ = np.zeros(stress_small_former_.shape[0]) # 1 for increase; 2 for decrease

threshold_increase_ =  np.percentile(stress_change_small_, percent_)
threshold_decrease_ =  np.percentile(stress_change_small_, 100 - percent_)

for i_ in range(stress_small_former_.shape[0]):
    if stress_change_small_[i_] > threshold_increase_:
        markers_small_[i_] = 1
  
    if stress_change_small_[i_] < threshold_decrease_:
        markers_small_[i_] = 2     
        
markers_small_ = markers_small_.astype('int')

# extract chain data

x_i_ = []
y_i_ = []
x_d_ = []
y_d_ = []

for i_ in range(stress_small_former_.shape[0]):
    if markers_small_[i_] == 1:  # increase
        x_i_.append(c_smaller_[i_][0])
        y_i_.append(c_smaller_[i_][1])   

    if markers_small_[i_] == 2:  # decrease
        x_d_.append(c_smaller_[i_][0])
        y_d_.append(c_smaller_[i_][1])   


x_i_ = np.array(x_i_)
y_i_ = np.array(y_i_)
x_d_ = np.array(x_d_)
y_d_ = np.array(y_d_)
```

```python
# plot the chain of small slip increase

plt.rc('font', family = 'Arial')
plt.rcParams['font.size'] = 8
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
fig = plt.figure(figsize=(7, 1))
plt.scatter(x_i_ - c_smaller_[:,0].min(),y_i_ - c_smaller_[:,1].min(), \
             s = 2, c = 'r')

# cb = plt.colorbar()
# cb. set_label(label = r'$\tau$' + ' (MPa)', fontsize = 8)

plt.xlabel('$\\mathit{x}$ (mm, from %d ms to %d ms)' % (time_4G_[events_4G_[n_small_][0][0]], time_4G_[events_4G_[n_small_][1][0]]), fontsize = 8)

plt.xlim([0, 360])
plt.ylabel('$\\mathit{y}$ (mm)',fontsize = 8)

plt.savefig('notebook_plot/IncreaseChain_SmallerSlip.svg', dpi=600, format='svg')
plt.show()

# plot the chain of small slip decrease

plt.rc('font', family = 'Arial')
plt.rcParams['font.size'] = 8
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
fig = plt.figure(figsize=(7, 1))
plt.scatter(x_d_ - c_smaller_[:,0].min(),y_d_ - c_smaller_[:,1].min(), \
             s = 2, c = 'b')

# cb = plt.colorbar()
# cb. set_label(label = r'$\tau$' + ' (MPa)', fontsize = 8)
plt.xlim([0, 360])
plt.xlabel('$\\mathit{x}$ (mm, from %d ms to %d ms)' % (time_4G_[events_4G_[n_small_][0][0]], time_4G_[events_4G_[n_small_][1][0]]), fontsize = 8)


plt.ylabel('$\\mathit{y}$ (mm)',fontsize = 8)

plt.savefig('notebook_plot/DecreaseChain_SmallerSlip.svg', dpi=600, format='svg')
plt.show()
```

```python
# get the stress of long slip

condition_ = '4G'

c_long_ = np.load('data/coordinate_data/c_former_of_%s_%s_from_%d_to_%d.npy' % \
                 (condition_, 'stick', events_4G_[n_long_-1][1][0], events_4G_[n_long_][0][0])   )
stress_long_former_ = np.load('data/stress_data/stresses_at_%d_%s.npy'% \
                        (events_4G_[n_long_ - 1][1][0], condition_) )
stress_long_later_ =  np.load('data/stress_data/stresses_at_%d_%s.npy'% \
                        (events_4G_[n_long_][0][0], condition_) )
tau_long_former_ = np.zeros(stress_long_former_.shape[0])
tau_long_later_ = np.zeros(stress_long_former_.shape[0])
# count_ = 0
for i_ in range(stress_long_former_.shape[0]):

    c_ = np.array([[stress_long_former_[i_][0], stress_long_former_[i_][2]],
              [stress_long_former_[i_][2], stress_long_former_[i_][1]]])
    eig, vector = np.linalg.eig(c_)
    maxarg_ = np.argmax(eig)
    minarg_ = np.argmin(eig)
    tau_long_former_[i_] = (eig[maxarg_] - eig[minarg_])/2
    
    c_ = np.array([[stress_long_later_[i_][0], stress_long_later_[i_][2]],
              [stress_long_later_[i_][2], stress_long_later_[i_][1]]])
    eig, vector = np.linalg.eig(c_)
    maxarg_ = np.argmax(eig)
    minarg_ = np.argmin(eig)
    tau_long_later_[i_] = (eig[maxarg_] - eig[minarg_])/2

stress_change_long_ =  tau_long_later_ - tau_long_former_
```

```python
# get the chain of long stick

# get the marker
markers_long_ = np.zeros(stress_long_former_.shape[0]) # 1 for increase; 2 for decrease

threshold_increase_ =  np.percentile(stress_change_long_, percent_)
threshold_decrease_ =  np.percentile(stress_change_long_, 100 - percent_)

for i_ in range(stress_long_former_.shape[0]):
    if stress_change_long_[i_] > threshold_increase_:
        markers_long_[i_] = 1
  
    if stress_change_long_[i_] < threshold_decrease_:
        markers_long_[i_] = 2     
        
markers_long_ = markers_long_.astype('int')

# extract chain data

x_i_ = []
y_i_ = []
x_d_ = []
y_d_ = []

for i_ in range(stress_long_former_.shape[0]):
    if markers_long_[i_] == 1:  # increase
        x_i_.append(c_long_[i_][0])
        y_i_.append(c_long_[i_][1])   

    if markers_long_[i_] == 2:  # decrease
        x_d_.append(c_long_[i_][0])
        y_d_.append(c_long_[i_][1])   


x_i_ = np.array(x_i_)
y_i_ = np.array(y_i_)
x_d_ = np.array(x_d_)
y_d_ = np.array(y_d_)
```
